```python
import os
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
from scipy.ndimage import gaussian_filter
import pandas as pd




def TotalProjection(matrix, bin_factor=1):
    N = matrix.shape[0]
    array = np.sum(matrix, axis=0)

    # Do NOT scale center shift — keep it fixed to match calibration
    center_shift = 12.5
    energy_axis = 1.785 + (0.99943 * (np.arange(N)*bin_factor))

    return plot_peaks_func(energy_axis, array, 'Energy', 'Counts', 'Total Projection', 'Energy (keV)', 'Counts', 'blue')

def plot_peaks_func(x_axis_data, y_axis_data, x_axis_name, y_axis_name, title_name,
                    x_axis_unit, y_axis_unit, color_type='blue'):
    # detect peaks (prominence at 1e2, should be changed)
    plot_peaks, _ = scipy.signal.find_peaks(y_axis_data, prominence=1e2)
    peak_energies = x_axis_data[plot_peaks]
    peak_counts = y_axis_data[plot_peaks]
    sorted_peaks = np.argsort(peak_counts)[::-1]

    # create df and plot
    spec_plot = pd.DataFrame({x_axis_name: x_axis_data, y_axis_name: y_axis_data})
    ax = spec_plot.plot(x=x_axis_name, y=y_axis_name, title=title_name,
                        xlabel=x_axis_unit, ylabel=y_axis_unit, color=color_type, figsize=(12, 6))

    # label peaks
    peak_list = []
    for i in sorted_peaks:
        x = peak_energies[i]
        peak_list.append(x)
        y = peak_counts[i]
        ax.plot(x, y, 'ro')  # red circle marker
        ax.text(x, y, f"{x:.1f} keV\n{int(y)}", fontsize=8, ha='center', va='bottom', rotation=0)

    peak_list = np.sort(np.round(peak_list, 1))
        

    plt.tight_layout()
    return peak_list

def TwoDPlot(matrix, file, filter_type):
    plt.imshow(matrix, cmap='inferno', norm=plt.matplotlib.colors.LogNorm(), origin='lower')
    plt.colorbar(label='Counts')
    plt.xlabel("Energy (keV)")
    plt.ylabel("Energy (keV)")
    plt.title(f"{file} with {filter_type} filter 2D Coincidence Plot")
    plt.show()
    return()
    


file_name = "m4bs/Eu152_at_Pb.m4b"
x = os.path.getsize(file_name)  # Should be 144000000 bytes
matrix = np.fromfile(file_name, dtype=np.uint32).reshape((4096, 4096))
print(x)


plt.imshow(matrix, cmap='inferno', norm=plt.matplotlib.colors.LogNorm(), origin='lower')
plt.colorbar(label='Counts')
plt.xlabel("Energy (keV)")
plt.ylabel("Energy (keV)")
plt.title("2D Coincidence Matrix from .m4b")
plt.show()

print(np.shape(matrix))
```

```python
folder_path = 'm4bs'
files = os.listdir(folder_path)

# 1st to last, excludes .ipynb_checkpoints folder
for file in files[1:]:
    file_name = f'm4bs/{file}'
    x = os.path.getsize(file_name)  # Should be 67108864 bytes
    matrix = np.fromfile(file_name, dtype=np.uint32).reshape((4096, 4096))
    raw_plot = TwoDPlot(matrix, file, 'no')

    # plotting
    median = scipy.ndimage.median_filter(matrix, size=51)
    gaussian = gaussian_filter(matrix, sigma=50)
    medain_plot = TwoDPlot(median, file, 'median')
    gaussian_plot = TwoDPlot(gaussian, file, 'gaussian')
    


    
    
    

    
```
